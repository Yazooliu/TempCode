## word2vec到sing2vec 

```python
#coding:utf-8 
"将歌曲的ID取出来，类似于分词完成的句子，送入word2vec中学习一下..."
import multiprocessing
import gensim  # NLP 的top model，涵盖word2vec 
import sys 
from random import shuffle

"解析从文件中读取的每一行的文本信息"
def parse_playlist_inline_to_get_sequence(in_line,playlist_sequence):
    "这里不必声明playlist_sequence 为list,多次声明，会使得返回playlist_sequence为[]"
    # playlist_sequence = []
    song_id_insequence = []
    contents_inline    = in_line.strip().split("\t")
    
    "解析每一行的歌单信息"
    for song in contents_inline[1:]:
        try:
            song_id, song_name, artist, popularity = song.split(":::")
            song_id_insequence.append(song_id)
        except:
            print "Exception message: song format error,please have a look..."
            print song + "\n"
            continue 
    
    "乱序处理 - 以保证每次取到的数据都是不一样"
    for i in range(len(song_id_insequence)):
        shuffle(song_id_insequence)
        playlist_sequence.append(song_id_insequence)


def song2vec(input_file,out_model):
    playlist_sequence = []
    for in_line in open(input_file):
        parse_playlist_inline_to_get_sequence(in_line,playlist_sequence)
    "获取CPU 计算核数目"
    cores = multiprocessing.cpu_count()
    print "Using all " + str(cores) + " CPU to training model...."

    model = gensim.models.Word2Vec(sentences=playlist_sequence, size=150, min_count=3, window=7, workers=cores)
    model.save(out_model)  
    print "Model Training Completed !"
```

```python
"训练song2vec模型"
in_file   = "./RawData/popular.playlist"  # 流行歌单和歌曲数据集合
out_model = "./Model/song2vec.model"
%time song2vec(in_file,out_model)
```

```python
"利用训练好的模型做做歌曲的预测"
import cPickle as pickle
popular_song_pload  =pickle.load(open("./InternalData/7_popular_song.pkl"))

"加载刚刚训练好的模型"
song2vec_model = gensim.models.Word2Vec.load("./Model/song2vec.model")

"一部分流行歌曲名字 - 可视化"
print "流行歌单中的前12个歌曲名字及歌手分别是:"
for popular_songid in popular_song_pload.keys()[0:12]:
    print popular_song_pload[popular_songid]
```

```python
"通过模型预测歌曲id - 根据一首歌预测10首歌"
listed_songid = popular_song_pload.keys()[2000:4000:50]
for songid in listed_songid:
    predicted_songid = song2vec_model.most_similar(songid)  # 返回[(预测Id,相似度)]
    
    print "跟歌曲: <<" + popular_song_pload[songid] + ">>,歌单ID:" + songid + "相似的歌曲和相似度是:"
    for songid in predicted_songid:
        perdictedsongid        = songid[0]
        perdictedsimilarscore  = songid[1]
        "通过预测的歌曲id找到预测的歌曲:"
        print popular_song_pload[perdictedsongid],"|",perdictedsimilarscore
        
    print "***********************************"
```

### "思考总结"

```python
"1,针对歌手发布的新歌或者收藏的冷门歌曲,如何处理？"
"方法一:  直接按照热门歌曲推荐，但是这种推荐并没有针对用户的个性化,并没有直接上解决问题"
"方法二:  从针对歌曲的推荐上升为针对歌手的推荐,针对用户喜好的歌手推荐类似风格的歌手给他，如果有新歌发布的歌手也可以做类似推荐.." 
"针对商品的推荐问题可以上升为针对商品品类的问题, 将品类list 送入gensim.models.Word2Vec 去训练。。。"

"2,针对用户兴趣问题,用户的喜好兴趣是变化的.前三年的喜好跟现在的喜好是不同是，是变化的。如何针对用户做推荐???"
   ". 每个人的喜好都具有时效性，可以按照用户收藏喜好的时间先后顺序做成时间轴，由近及远标上不同的衰减因子，最近一首歌是1，前一首歌是0.98..."
   ". 针对不同的歌曲的热度来排序，最热度100是的权重因子是1，热度98的权重因子是0.99"

"最后对每一首歌都可以那会一个song2vec的推荐列表和相应的相似度,相似度可以通过时间衰减因子和热度权重来做加权求和，然后排序后展示给用户"
```
