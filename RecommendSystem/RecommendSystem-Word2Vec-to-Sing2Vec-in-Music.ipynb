## word2vec到sing2vec 

```python
"将歌曲的ID取出来，类似于分词完成的句子，送入word2vec中学习一下..."
#coding:utf-8 
import multiprocessing
import gensim  # NLP 的top model，涵盖word2vec 
import sys 
from random import shuffle

"解析从文件中读取的每一行的文本信息"
def parse_playlist_inline_to_get_sequence(in_line,playlist_sequence):
    "input : in_line"
    "output: playlist_sequence "
    song_id_insequence = []
    contents_inline    = in_line.strip().split("\t")
    
    "解析每一行的歌单信息"
    for _inline in contents_inline[1:]:
        try:
            song_id,song_name,song_artist,song_popularityscores = _inline.split(":::")
            song_id_insequence.append(song_id)
        except:
            print "Exception infor: song format error... "
            print _inline + "\n"
            continue 
    
    "乱序处理 - 以保证每次取到的数据都是不一样"
    playlist_sequence = []
    for i in range( len(song_id_insequence) ):
        shuffle(song_id_insequence)
        playlist_sequence.append(song_id_insequence)
    
    return playlist_sequence


def song2vec(input_file):
    "Defination"
    playlist_sequence = []
    for in_line in open(input_file):
        parse_playlist_inline_to_get_sequence(in_line,playlist_sequence)
        
    
    "获取CPU 计算核数目"
    cpu_cores = multiprocessing.cpu_count()
    print "Using all " + cpu_cores + "to training model...."
    "Call Gensim Word2Vec Model to train playlist_sequence "
    "playlist_sequence - is song id in sequence "
    from gensim.models import Word2Vec
    model = Word2Vec(common_texts = playlist_sequence, size=100, window=5, min_count=1, workers=cpu_cores) 
    model.save("./Model/song2vec.model")  # save model to local
    print "Model Training Completed !"
    
"训练song2vec 模型"
%time song2vec()
```
